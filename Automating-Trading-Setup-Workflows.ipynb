```python
######################################
# Generate Trading Alert
######################################

def calculate_total_pnl(close_prices, period_days):
    """
    Calculates the total percentage profit/loss over a given period as specified by user.
    """
    # Calculate the price at the start of the period
    price_start_of_period = close_prices.shift(period_days)
    # Calculate total PnL%
    total_pnl_percent = ((close_prices / price_start_of_period) - 1) * 100
    return total_pnl_percent


def calculate_rsi(close_prices, period):
    """
    Calculates the RSI (Relative Strength Index) over a given period as specified by user.
    """

    delta = close_prices.diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)

    avg_gain = gain.rolling(window=period, min_periods=period).mean()
    avg_loss = loss.rolling(window=period, min_periods=period).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

def calculate_volatility(close_prices, period, annualization_factor=252):
    daily_returns = close_prices.pct_change()
    volatility = daily_returns.rolling(window=period).std() * (annualization_factor**0.5)
    return volatility

def rolling_sharpe_ratio(returns_series, window_days, annualization_factor_sqrt=252**0.5):
    """
    Calculates the annualized rolling Sharpe Ratio for a given period as specified by user.
    Assumes a risk-free rate of 0.
    """
    # Calculate rolling mean and standard deviation of daily returns
    rolling_mean_returns = returns_series.rolling(window=window_days).mean()
    rolling_std_returns = returns_series.rolling(window=window_days).std()

    # Avoid division by zero for cases where standard deviation is zero
    # Fill NaN from std=0 with 0 Sharpe, as there's no risk or return variability
    sharpe = (rolling_mean_returns / rolling_std_returns).fillna(0)
    return sharpe * annualization_factor_sqrt # Annualize the Sharpe Ratio

def calculate_macd(close_prices, fast_period=12, slow_period=26, signal_period=9):
    """
    Calculates MACD (Moving Average Convergence Divergence) indicators.
    """
    exp1 = close_prices.ewm(span=fast_period, adjust=False).mean()
    exp2 = close_prices.ewm(span=slow_period, adjust=False).mean()
    macd_line = exp1 - exp2
    signal_line = macd_line.ewm(span=signal_period, adjust=False).mean()
    macd_histogram = macd_line - signal_line
    return macd_line, signal_line, macd_histogram

def calculate_bollinger_bands(close_prices, window=200, num_std_dev=2):
    """
    Calculates Bollinger Bands (Middle, Upper, Lower).
    """
    # Calculate the Middle Band (SMA)
    middle_band = close_prices.rolling(window=window).mean()

    # Calculate the standard deviation
    std_dev = close_prices.rolling(window=window).std()

    # Calculate Upper and Lower Bands
    upper_band = middle_band + (std_dev * num_std_dev)
    lower_band = middle_band - (std_dev * num_std_dev)

    return middle_band, upper_band, lower_band
```

```python
# Calculate Technical Indicators

# Prepare a new DataFrame to store results with MultiIndex columns
# The columns will be (Ticker, Metric)
features_data = pd.DataFrame(index=data.index)

for ticker in tickers:
    print("Running: ", ticker)

    try:
      # Directly extract 'Adj Close' prices for the current ticker from the MultiIndex DataFrame
      # Using .loc for explicit access to the column level
      close_prices = data.loc[:, (ticker, 'Adj Close')].dropna() # dropna here to handle potential leading NaNs from yfinance
      high_prices = data.loc[:, (ticker, 'High')].dropna() # Get High prices
      open_prices = data.loc[:, (ticker, 'Open')].dropna() # Get Open prices
      low_prices = data.loc[:, (ticker, 'Low')].dropna() # Get Open prices

      ###############################################
      # 1. Add Adjusted Close Price
      ###############################################
      features_data[(ticker, f'ticker')] = ticker
      features_data[(ticker, 'Adj Close')] = close_prices
      features_data[(ticker, 'Open')] = open_prices
      features_data[(ticker, 'Low')] = low_prices

      ###############################################
      # New: Add Highest Price Today and Highest Price Today * 99%
      ###############################################
      features_data[(ticker, 'Highest Price Today')] = high_prices
      features_data[(ticker, 'Highest Price Today * 0.99')] = high_prices * 0.99

      ###############################################
      # 2. Calculate Daily PNL %
      ###############################################
      features_data[(ticker, 'Daily_PNL%')] = close_prices.pct_change() * 100
      features_data[(ticker, 'Daily_PNL_1scale')] = close_prices.pct_change() + 1
      # features_data[(ticker, 'PnL_P14D')] =   calculate_total_pnl(close_prices, p14d_days)
      # features_data[(ticker, 'PnL_P1M')] =   calculate_total_pnl(close_prices, p1m_days)
      # features_data[(ticker, 'PnL_P3M')] =   calculate_total_pnl(close_prices, p3m_days)
      # features_data[(ticker, 'PnL_P6M')] =   calculate_total_pnl(close_prices, p6m_days)
      # features_data[(ticker, 'PnL_P12M')] =   calculate_total_pnl(close_prices, p12m_days)

      ###############################################
      # 3. Calculate RSI (Relative Strength Index)
      ###############################################
      features_data[(ticker, 'RSI_P14D')] = calculate_rsi(close_prices, p14d_days)
      # features_data[(ticker, 'RSI_P1M')] = calculate_rsi(close_prices, p1m_days)
      # features_data[(ticker, 'RSI_P3M')] = calculate_rsi(close_prices, p3m_days)
      # features_data[(ticker, 'RSI_P6M')] = calculate_rsi(close_prices, p6m_days)
      # features_data[(ticker, 'RSI_P12M')] = calculate_rsi(close_prices, p12m_days)

      ###############################################
      # 4. Calculate Volatility (252-day annual historical volatility)
      ###############################################
      # features_data[(ticker, 'Vol_P14D')] = calculate_volatility(close_prices, p14d_days)
      # features_data[(ticker, 'Vol_P1M')] = calculate_volatility(close_prices, p1m_days)
      # features_data[(ticker, 'Vol_P3M')] = calculate_volatility(close_prices, p3m_days)
      # features_data[(ticker, 'Vol_P6M')] = calculate_volatility(close_prices, p6m_days)
      features_data[(ticker, 'Vol_P12M')] = calculate_volatility(close_prices, p12m_days)

      ###############################################
      # 5. Calculate Moving Averages
      ###############################################
      # features_data[(ticker, 'MA_P14D')] = close_prices.rolling(window=p14d_days).mean()
      features_data[(ticker, 'MA_P20D')] = close_prices.rolling(window=20).mean() # ~20 day MA
      features_data[(ticker, 'MA_P50D')] = close_prices.rolling(window=50).mean() # ~50 day MA
      # features_data[(ticker, 'MA_P1M')] = close_prices.rolling(window=p1m_days).mean() # ~1 month
      # features_data[(ticker, 'MA_P3M')] = close_prices.rolling(window=p3m_days).mean() # ~3 months
      # features_data[(ticker, 'MA_P6M')] = close_prices.rolling(window=p6m_days).mean() # ~6 months
      features_data[(ticker, 'MA_P12M')] = close_prices.rolling(window=p12m_days).mean()
      features_data[(ticker, 'MA_200D')] = close_prices.rolling(window=200).mean()

      ###############################################
      # 6. Number of months where current Adj Close > monthly min (past 12 months)
      ###############################################
      # Calculate monthly minimums for the entire price history for easier lookup
      monthly_mins_df = close_prices.resample('ME').min()
      higher_than_monthly_min_counts = []

      for current_date, current_price in close_prices.items():
          count = 0
          # Iterate over the last 12 calendar months relative to the current_date
          for i in range(1, 13):
              # Calculate the end date of the month 'i' months ago
              # Using day=31 to ensure we correctly identify the relevant month-end for comparison
              target_month_end = current_date - pd.DateOffset(months=i, day=31)

              # Find the monthly min that corresponds to this target month
              # Use searchsorted to find the index where target_month_end would be inserted
              # And then take the preceding index for the actual month-end from monthly_mins_df
              idx_pos = monthly_mins_df.index.searchsorted(target_month_end, side='right') - 1

              if idx_pos >= 0:
                  relevant_monthly_min_date = monthly_mins_df.index[idx_pos]
                  monthly_min_value = monthly_mins_df.loc[relevant_monthly_min_date]
                  # Check for NaN values before comparison
                  if pd.notna(current_price) and pd.notna(monthly_min_value) and current_price > monthly_min_value:
                      count += 1
          higher_than_monthly_min_counts.append(count)

      features_data[(ticker, 'Higher_Than_Monthly_Min_Count_P12M')] = pd.Series(higher_than_monthly_min_counts, index=close_prices.index)

      ###############################################
      # 7. MA Crossover Flags
      ###############################################
      # Bullish Crossover: 20MA crosses above 50MA
      features_data[(ticker, 'MA_20_50_Bullish_Crossover')] = ((features_data[(ticker, 'MA_P20D')] > features_data[(ticker, 'MA_P50D')]) &
                                                              (features_data[(ticker, 'MA_P20D')].shift(1) <= features_data[(ticker, 'MA_P50D')].shift(1))).astype(int)
      # Bearish Crossover: 50MA crosses above 20MA (or 20MA crosses below 50MA)
      features_data[(ticker, 'MA_50_20_Bearish_Crossover')] = ((features_data[(ticker, 'MA_P50D')] > features_data[(ticker, 'MA_P20D')]) &
                                                              (features_data[(ticker, 'MA_P50D')].shift(1) <= features_data[(ticker, 'MA_P20D')].shift(1))).astype(int)

      ###############################################
      # 8. MA 20D Higher Than 50D Flag
      ###############################################
      features_data[(ticker, 'MA_20_Higher_Than_50_Flag')] = (features_data[(ticker, 'MA_P20D')] > features_data[(ticker, 'MA_P50D')]).astype(int)

      ###############################################
      # 9. Sharpe Ratio
      ###############################################
      daily_returns = close_prices.pct_change()
      # features_data[(ticker, 'Sharpe_Ratio_P1M')] = rolling_sharpe_ratio(daily_returns, p1m_days)
      # features_data[(ticker, 'Sharpe_Ratio_P3M')] = rolling_sharpe_ratio(daily_returns, p3m_days)
      # features_data[(ticker, 'Sharpe_Ratio_P6M')] = rolling_sharpe_ratio(daily_returns, p6m_days)
      features_data[(ticker, 'Sharpe_Ratio_P12M')] = rolling_sharpe_ratio(daily_returns, p12m_days)

      ###############################################
      # 10. MACD
      ###############################################
      macd_line, signal_line, macd_histogram = calculate_macd(close_prices)
      features_data[(ticker, 'MACD_Line')] = macd_line
      features_data[(ticker, 'Signal_Line')] = signal_line
      features_data[(ticker, 'MACD_Histogram')] = macd_histogram

      ###############################################
      # 11. New MACD Indicators
      ###############################################
      # Bullish Crossover of MACD Line over Signal Line
      features_data[(ticker, 'MACD_Signal_Bullish_Crossover')] = ((macd_line > signal_line) &
                                                                  (macd_line.shift(1) <= signal_line.shift(1)) ).astype(int)

      # # Bearish Crossover of MACD Line over Signal Line
      # features_data[(ticker, 'MACD_Signal_Bullish_Crossover')] = ((macd_line > signal_line) &
      #                                                             (macd_line.shift(1) <= signal_line.shift(1))).astype(int)

      # MACD Line Crossover from Negative to Positive (Zero Line Bullish Crossover)
      features_data[(ticker, 'MACD_Signal_Bearish_Crossover')] = ((macd_line < signal_line ) &
                                                                      (macd_line.shift(1) >= signal_line.shift(1))  ).astype(int)

      ###############################################
      # 12. Bollinger Bands (200)
      ###############################################
      window_bb = 200
      num_std_dev = 2 # K=2 is standard

      middle_band, upper_band, lower_band = calculate_bollinger_bands(close_prices, window=window_bb, num_std_dev=num_std_dev)

      features_data[(ticker, f'BB_Middle_Band_P{window_bb}D')] = middle_band
      features_data[(ticker, f'BB_Upper_Band_P{window_bb}D')] = upper_band
      features_data[(ticker, f'BB_Lower_Band_P{window_bb}D')] = lower_band

      ###############################################
      # 13. Bollinger Band Price Position Flags
      ###############################################
      features_data[(ticker, 'Price_Above_Upper_BB')] = (close_prices > features_data[(ticker, f'BB_Upper_Band_P{window_bb}D')]).astype(int)
      features_data[(ticker, 'Price_Below_Lower_BB')] = (close_prices < features_data[(ticker, f'BB_Lower_Band_P{window_bb}D')]).astype(int)

      features_data[(ticker, 'Candle_Above_Upper_BB')] = (high_prices >= features_data[(ticker, f'BB_Upper_Band_P{window_bb}D')]).astype(int)
      features_data[(ticker, 'Candle_Below_Lower_BB')] = (low_prices <= features_data[(ticker, f'BB_Lower_Band_P{window_bb}D')]).astype(int)



      print("Finished: ", ticker)


    except:
      print("Ticker not retrieved")
```

```python
import pandas as pd

# Convert the current tuple-based column names to a proper MultiIndex
# This is crucial for stacking to work correctly by level name
features_data.columns = pd.MultiIndex.from_tuples(features_data.columns, names=['Ticker_Symbol', 'Metric'])

# Stack the DataFrame, moving the 'Ticker_Symbol' level from columns to rows
# This transforms the DataFrame from wide to long format
# The result will have a MultiIndex (original_index, Ticker_Symbol) and metrics as columns
features_data_long = features_data.stack(level='Ticker_Symbol')

# Reset the index to convert the MultiIndex (Date, Ticker_Symbol) into regular columns
# 'level_0' will be 'Date' (from original index), 'level_1' will be 'Ticker_Symbol'
features_data_long = features_data_long.reset_index()

# Rename the columns for clarity
features_data_long = features_data_long.rename(columns={'level_0': 'Date', 'level_1': 'Ticker'})

# The original MultiIndex also had a ('Ticker_Symbol', 'ticker') column which contained the ticker symbol.
# When stacked, this creates a column named 'ticker' which is redundant as we now have a 'Ticker' column.
if 'ticker' in features_data_long.columns:
    features_data_long = features_data_long.drop(columns=['ticker'])

# Display the first few rows of the transformed DataFrame
# print(features_data_long.head())

# Update the features_data variable to reflect the new long format
features_data = features_data_long
```
